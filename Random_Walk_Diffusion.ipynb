```python
import porespy as ps
import pytrax as pt
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
# A = ps.generators.blobs([200,200,200],porosity=0.4,blobiness=1)
A = ps.generators.polydisperse_spheres([200,200,200], 0.4, dist = scipy.stats.norm(loc=20, scale=10))
B = np.swapaxes(A,axis1=1, axis2=2)
fig, ax = plt.subplots(1,2)
ax[0].imshow(A[0,:,:])
ax[1].imshow(B[0,:,:])
plt.show()
```

```python
rwA = pt.RandomWalk(A)
rwB = pt.RandomWalk(B)
rwA.run(5000, 5000)
rwB.run(5000, 5000)
```

```python
rwA.calc_msd()
rwA.plot_msd()
plt.show()
```

```python
rwB.calc_msd()
rwB.plot_msd()
plt.show()
```
